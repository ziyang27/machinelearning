```python
# Import libraries
import pandas as pd
import seaborn as sns
```

```python
# Read data
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
# Inspect the data
print(df.head())
print(df.shape)
print(df.dtypes)
```

```python
# Clean up the data
# Only want 50 km or 50 mile, 2020
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]  
```

```python
# Only want USA events
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']  
```

```python
# Combine all filters together
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]  
```

```python
# Inspect again
print(df2.head(10))
print(df2.shape)
```

```python
# Remove USA from event name
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
# Inspect again
df2.head(10)
```

```python
# Create new athlete age feature
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
# Inspect again
df2.head(10)
```

```python
# Clean up the format of athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
# Drop columns: Athlete club, Athlete country, Athlete YOB, Athlete age category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
# Inspect again
df2.head()
```

```python
# Clean up null values
df2.isna().sum()
```

```python
df2 = df2.dropna()
df2.isna().sum()
```

```python
# Check for duplicate values
df2[df2.duplicated() == True]
```

```python
# Reset index
df2.reset_index(drop = True)
```

```python
# Change athelete age to integer
df2['Athlete age'] = df2['Athlete age'].astype(int)
```

```python
# Change athlete performance to float
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# Inspect again
df2.head(10)
```

```python
# Rename columns
df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_date',
    'Event name': 'race_name',           
    'Event distance/length': 'race_length',     
    'Event number of finishers': 'race_no_of_finishers',      
    'Athlete performance': 'athlete_performance',           
    'Athlete gender': 'athlete_gender',                
    'Athlete average speed': 'athlete_avg_speed',       
    'Athlete ID': 'athlete_id',                     
    'Athlete age': 'athlete_age'  
})
```

```python
# Reorder columns
df3 = df2[['race_date', 'race_name', 'race_length', 'race_no_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_avg_speed']]
```

```python
# Create a histogram (more of a bar chart)
sns.histplot(df3['race_length'])
```

```python
# Create a histogram with hue (more of a bar chart)
sns.histplot(df3, x = 'race_length', hue = 'athlete_gender')
```

```python
# Create a histogram
sns.displot(df3[df3['race_length'] == '50mi']['athlete_avg_speed'])
```

```python
# Create a violin plot
sns.violinplot(data = df3, x = 'race_length', y = 'athlete_avg_speed', hue = 'athlete_gender', split = True, inner = 'quart')
```

```python
# Create a scatter plot with regression line
sns.lmplot(data=df3, x='athlete_age', y='athlete_avg_speed', hue='athlete_gender')
```

```python
# Get the median athlete average speed
df3.groupby(['race_length', 'athlete_gender'])['athlete_avg_speed'].median()
```

```python
# Query for the top 15 athlete ages with the highest average
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19').head(15)
```

```python
# Create a new feature : race_month
df3['race_month'] = df3['race_date'].str.split('.').str.get(1).astype(int)
```

```python
# Create a new feature: race_season
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
# Inspect again
df3.head(10)
```

```python
# Get the average speed by season
df3.groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean')
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean')
```
